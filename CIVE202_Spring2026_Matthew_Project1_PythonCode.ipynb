```python
import pandas as pd
```

```python
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')
air_data = pd.DataFrame(air_data)
air_data
```

```python
air_data_pm25 = air_data.groupby(['sensor.name']).agg(mean =('pm2.5_atm', 'mean'),
                                                     median =('pm2.5_atm', 'median'))
air_data_pm25
```

```python
top5_pm25 = air_data_pm25.sort_values(by = 'mean', ascending = False)
top5_pm25.head(5)
```

```python
air_data_pm10 = air_data.groupby(['sensor.name']).agg(mean =('pm10.0_atm', 'mean'),
                                                     median =('pm10.0_atm', 'median'))
air_data_pm10
```

```python
top5_pm10 = air_data_pm10.sort_values(by = 'mean', ascending = False)
top5_pm10.head(5)
```

```python
air_data_voc = air_data.groupby(['sensor.name']).agg(mean =('voc', 'mean'),
                                                     median =('voc', 'median'))
air_data_voc
```

```python
top5_voc = air_data_voc.sort_values(by = 'mean', ascending = False)
top5_voc.head(5)
```

```python
def assign_temperature(X): # this function takes only one variable designated as X
    if X < 32:
        return 'Below Freezing'
    elif 32 <= X <= 50:
        return 'Cool'
    elif 51 <= X <= 70:
        return 'Warm'
    else:
        return 'Hot'    # This means Hot is greater than >71
```

```python
air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)

air_data
```

```python
def assign_humidity(Y):
    if Y < 50:
        return 'Low Humidity'
    elif 50 <= Y <= 80:
        return 'High Humidity'
    else:
        return 'Very High Humidity'
```

```python
air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)

air_data
```

```python
# Use the maximum values on location and date

unhealthy_pm25 = air_data[ air_data['pm2.5_atm'] >= 50 ] # taking on rows of data with pm2.5 >= 25

unhealthy_dates = unhealthy_pm25.sort_values('date')
unhealthy_dates
```

```python
unhealthy_pm10 = air_data[ air_data['pm10.0_atm'] >= 50 ] # taking on rows of data with pm10 >= 50

unhealthy_dates = unhealthy_pm10.sort_values('date')
unhealthy_dates
```

```python
unhealthy_voc = air_data[ air_data['voc'] >= 150 ] # taking on rows of data with voc >= 150

unhealthy_dates = unhealthy_voc.sort_values('date')
unhealthy_dates
```
